# Brown Corpus
- Statistics
  - ~1.1 Million tokens
  - 500 text documents
  - 15 categories

```python
import nltk
from nltk.corpus import brown
```

```python
tokens = brown.words()
```

```python
print(f'TTR of brown corpus: {len(set(tokens))/len(tokens)}')
```

```python
tokenCount = nltk.FreqDist(tokens)
print(tokenCount)
```

```python
print(tokenCount.most_common(50))
```

```python
tokenCount.plot(50, cumulative=True)
```

```python
print(f'Tokens in corpus: {len(tokens)}')
```

## Frequency of frequency
- How many times words appeared n times

```python
from collections import defaultdict
countFreq = defaultdict(int)
for token in tokenCount:
	countFreq[tokenCount[token]] += 1
print(countFreq)
```

```python
import matplotlib.pyplot as plt
plt.scatter(list(countFreq.keys()), list(countFreq.values()))
plt.xlabel("Tokens Count")
plt.xlabel("Frequency of token count")
plt.xscale("log")
plt.yscale("log")
plt.show()
```
